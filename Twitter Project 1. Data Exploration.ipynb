**Checking the "User" column which has we can see it contains another dictionary in it.**

````python
df_ev['user']
````

**Using this function in order to be able to read the dictionary inside the column.**

````python
def expand_dictionary(row, columnname=''):
    if columnname == '':
        return row
    # Returning the original row if the column does not contain a dictionary
    if type(row[columnname]) != dict:
        return row
    
    # Otherwise, transforming the dictionary in additional columns
    for key, value in row[columnname].items():
        if type(value) != dict:
            row[key] = value
        else:
            for subkey, subvalue in row[columnname][key].items():
                if type(subvalue) != dict:
                    row[str(key)+'_'+str(subkey)] = subvalue
                else:
                    for subsubkey, subsubvalue in row[columnname][key][subkey].items():
                        row[str(key)+'_'+str(subkey)+'_'+str(subsubkey)] = subsubvalue
                
        
    return row 
````

**Creating a new dataset which contains the additional dictionary present in the "User" columns.**

````python
df_ev2 = df_ev.apply(expand_dictionary, axis=1, args=('user',))
````

**Checking which additional columns the new dictionary was containing.**

````python
df_ev2.columns
````

**After finding the columns we needed, "followers_count", we can check the minimum, the average and the max followers the Users of this Dataset have aggregating this values together.**

````python
df_ev2['followers_count'].agg(['mean','min','max'])
````

**Otherwise we can run a simple command as such the one below and see that the average number of followers of our users is 29,5K, the minumum is 0 and the max is 8,5 Million followers. Task 7 completed.**

````python
df_ev2['followers_count'].describe()
````
